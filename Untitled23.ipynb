```python
# --- 1. Install ALL required packages ---
!pip install torch stable-baselines3 pandas numpy matplotlib scikit-learn 'shimmy>=2.0' requests

# --- 2. Upload & setup SCAPS file ---
import shutil, numpy as np
template_path = '/content/CIGSfirst point (2).scaps'
working_path = '/content/working_scaps.scaps'
shutil.copy(template_path, working_path)

# --- 3. Download NASA POWER weather data (hourly) ---
import pandas as pd, requests
latitude, longitude = 53.524960, -1.627447
start, end = '20200101', '20201231'
url = (
    "https://power.larc.nasa.gov/api/temporal/hourly/point"
    f"?parameters=ALLSKY_SFC_SW_DWN,T2M,WS2M,RH2M,CLRSKY_SFC_SW_DWN"
    f"&community=RE"
    f"&longitude={longitude}&latitude={latitude}"
    f"&start={start}&end={end}&format=CSV"
)
response = requests.get(url)
with open('/content/nasa_power.csv', 'wb') as f: f.write(response.content)
# Preview header; adjust skiprows below as needed based on printout
with open('/content/nasa_power.csv') as f:
    for i in range(25): print(f.readline())
weather_df = pd.read_csv('/content/nasa_power.csv', skiprows=14)
weather_df = weather_df.reset_index(drop=True)
print(weather_df.head())

# --- 4. Define SCAPS file update function for multi-layer structure ---
def update_scaps_file(file_path, params):
    with open(file_path, "r") as f:
        lines = f.readlines()
    for i, line in enumerate(lines):
        if "CIGS_thickness" in line and 'CIGS_thickness' in params:
            lines[i] = f"CIGS_thickness = {params['CIGS_thickness']}\n"
        if "CdS_thickness" in line and 'CdS_thickness' in params:
            lines[i] = f"CdS_thickness = {params['CdS_thickness']}\n"
        if "ZnO_thickness" in line and 'ZnO_thickness' in params:
            lines[i] = f"ZnO_thickness = {params['ZnO_thickness']}\n"
        if "Mo_thickness" in line and 'Mo_thickness' in params:
            lines[i] = f"Mo_thickness = {params['Mo_thickness']}\n"
    with open(file_path, "w") as f:
        f.writelines(lines)

# --- 5. Define surrogate SCAPS simulator (replace with actual SCAPS batch call for real Î·) ---
def run_scaps_simulation(file_path):
    simulated_eta = np.random.uniform(0.12, 0.20)
    return simulated_eta

# --- 6. RL Environment connecting weather, SCAPS, and RL action ---
import gym
from gym import spaces

class SolarCellEnv(gym.Env):
    def __init__(self):
        super().__init__()
        self.action_space = spaces.Box(
            low=np.array([0.1, 0.05, 0.05, 0.05]),   # min thickness for CIGS, CdS, ZnO, Mo (Î¼m)
            high=np.array([4.0, 0.5, 0.5, 1.0]),     # max thickness
            dtype=np.float32)
        self.observation_space = spaces.Box(
            low=-np.inf, high=np.inf, shape=(weather_df.shape[1],), dtype=np.float32)
        self.current_step = 0
        self.max_steps = weather_df.shape[0]

    def reset(self, *, seed=None, options=None):
        self.current_step = 0
        obs = weather_df.iloc[self.current_step].values.astype(np.float32)
        info = {}
        return obs, info

    def step(self, action):
        params = {
            'CIGS_thickness': float(action[0]),
            'CdS_thickness': float(action[1]),
            'ZnO_thickness': float(action[2]),
            'Mo_thickness': float(action[3])
        }
        update_scaps_file(working_path, params)
        eta = run_scaps_simulation(working_path)
        obs = weather_df.iloc[self.current_step].values.astype(np.float32)
        reward = eta
        self.current_step += 1
        done = self.current_step >= self.max_steps
        info = {"eta": eta, "params": params}
        return obs, reward, done, False, info

# --- 7. Train RL PPO agent to maximize Î· ---
from stable_baselines3 import PPO
env = SolarCellEnv()
model = PPO("MlpPolicy", env, verbose=1, learning_rate=1e-3)
model.learn(total_timesteps=30000)
model.save("ppo_solarcell")

# --- 8. Evaluate and plot optimized Î· ---
import matplotlib.pyplot as plt
obs, info = env.reset()
etas, params_list = [], []
for _ in range(env.max_steps):
    action, _ = model.predict(obs)
    obs, reward, done, _, info = env.step(action)
    etas.append(info['eta'])
    params_list.append(info['params'])
    if done: break
max_eta = max(etas)
best_params = params_list[etas.index(max_eta)]
print("Optimized Î·:", max_eta)
print("Best device parameters:", best_params)
plt.plot(etas)
plt.xlabel('Time-step')
plt.ylabel('Efficiency Î·')
plt.title('Optimized Solar Cell Efficiency Over Weather')
plt.show()
```
